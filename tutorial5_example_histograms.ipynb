### Tutorial: Histogram

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
sugar_level = [93, 145, 124, 87, 97, 108, 112, 130, 119, 95, 105]
```

```python
plt.hist(sugar_level)
```

```python
plt.hist(sugar_level, bins=3)
```

```python
plt.hist(sugar_level, bins=3, rwidth=0.95)
```

```python
plt.hist(sugar_level, bins=[80, 100, 125, 150], rwidth=0.95)
```

```python
plt.hist(sugar_level, bins=[80, 100, 125, 150], rwidth=0.95, histtype='step')
```

```python
plt.hist(sugar_level, bins=[80, 100, 125, 150], rwidth=0.95, color='green')
```

```python
plt.hist(sugar_level, bins=[80, 100, 125, 150], rwidth=0.95, color='g')
```

```python
sugar_men = [89, 111, 139, 94, 84, 94, 130, 126, 99]
sugar_women = [109, 93, 89, 107, 149, 109, 129, 141]
```

```python
plt.hist([sugar_men, sugar_women], bins=[80, 100, 125, 150], rwidth=0.95, color=['g', 'blue'])
```

```python
plt.title('Sugar report')
plt.xlabel('Sugar range')
plt.ylabel('no of patients')
plt.hist([sugar_men, sugar_women], bins=[80, 100, 125, 150], rwidth=0.95, color=['g', 'blue'], label=['men', 'women'])
plt.legend()
```

```python
plt.title('Sugar report')
plt.xlabel('no of patients')
plt.ylabel('Sugar level range')
plt.hist([sugar_men, sugar_women], bins=[80, 100, 125, 150], rwidth=0.95, color=['g', 'blue'], label=['men', 'women'], orientation='horizontal')
plt.legend()
```
