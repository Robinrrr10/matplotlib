### Tutorial 6 - Pie chart

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
home_expenses = [1500, 700, 1200, 500, 800]
home_expense_labels = ['groccery', 'travel', 'rent', 'internet/phone', 'electricity']
```

```python
plt.pie(home_expenses, labels=home_expense_labels)
```

```python
plt.axis('equal')
plt.pie(home_expenses, labels=home_expense_labels)
```

```python
plt.pie(home_expenses, labels=home_expense_labels, radius=2)
```

```python
plt.axis('equal')
plt.pie(home_expenses, labels=home_expense_labels, radius=1.5)
```

```python
plt.axis('equal')
plt.pie(home_expenses, labels=home_expense_labels, radius=1, autopct='%0.2f%%')
```

```python
plt.pie(home_expenses, labels=home_expense_labels, radius=1, autopct='%0.2f%%', shadow=True)
```

```python
plt.pie(home_expenses, labels=home_expense_labels, radius=1, autopct='%0.2f%%', explode=[0, 0, 0.2, 0, 0])
```

```python
plt.pie(home_expenses, labels=home_expense_labels, radius=1, autopct='%0.2f%%', explode=[0, 0, 0.2, 0, 0.1])
```

```python
plt.pie(home_expenses, labels=home_expense_labels, radius=1, autopct='%0.2f%%', startangle=45)
```
