### Tutorial 3 : label, legend and grid

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
days = [1, 2, 3, 4, 5, 6, 7]
temp_avg = [39, 40, 42, 41, 42, 43, 41]
temp_min = [37, 38, 38, 39, 40, 40, 38]
temp_max = [41, 43, 44, 42, 43, 45, 43]
```

```python
plt.title('Weather')
plt.xlabel('Days')
plt.ylabel('Temperature')
plt.plot(days, temp_avg)
plt.plot(days, temp_min)
plt.plot(days, temp_max)
```

```python
plt.title('Weather')
plt.xlabel('Days')
plt.ylabel('Temperature')
plt.plot(days, temp_avg, label = 'Avg')
plt.plot(days, temp_min, label = 'Min')
plt.plot(days, temp_max, label = 'Max')
plt.legend()
```

```python
plt.title('Weather')
plt.xlabel('Days')
plt.ylabel('Temperature')
plt.plot(days, temp_avg, label = 'Avg')
plt.plot(days, temp_min, label = 'Min')
plt.plot(days, temp_max, label = 'Max')
plt.legend(loc = 'upper right')
```

```python
plt.title('Weather')
plt.xlabel('Days')
plt.ylabel('Temperature')
plt.plot(days, temp_avg, label = 'Avg')
plt.plot(days, temp_min, label = 'Min')
plt.plot(days, temp_max, label = 'Max')
plt.legend(loc = 'best')
```

```python
plt.title('Weather')
plt.xlabel('Days')
plt.ylabel('Temperature')
plt.plot(days, temp_avg, label = 'Avg')
plt.plot(days, temp_min, label = 'Min')
plt.plot(days, temp_max, label = 'Max')
plt.legend(loc = 'best', shadow=True)
```

```python
plt.title('Weather')
plt.xlabel('Days')
plt.ylabel('Temperature')
plt.plot(days, temp_avg, label = 'Avg')
plt.plot(days, temp_min, label = 'Min')
plt.plot(days, temp_max, label = 'Max')
plt.legend(loc = 'best', shadow=True, fontsize='large')
```

```python
plt.title('Weather')
plt.xlabel('Days')
plt.ylabel('Temperature')
plt.plot(days, temp_avg, label = 'Avg')
plt.plot(days, temp_min, label = 'Min')
plt.plot(days, temp_max, label = 'Max')
plt.legend(loc = 'best', shadow=True, fontsize='large')
```

```python
plt.title('Weather')
plt.xlabel('Days')
plt.ylabel('Temperature')
plt.plot(days, temp_avg, label = 'Avg')
plt.plot(days, temp_min, label = 'Min')
plt.plot(days, temp_max, label = 'Max')
plt.legend(loc = 'best', shadow=True, fontsize='small')
```

```python
plt.title('Weather')
plt.xlabel('Days')
plt.ylabel('Temperature')
plt.plot(days, temp_avg, label = 'Avg')
plt.plot(days, temp_min, label = 'Min')
plt.plot(days, temp_max, label = 'Max')
plt.legend(loc = 'best', shadow=True, fontsize='small')
plt.grid()
```

```python
plt.title('Weather')
plt.xlabel('Days')
plt.ylabel('Temperature')
plt.plot(days, temp_avg, label = 'Avg', marker='*', linewidth=5, markersize=10, linestyle='dotted')
plt.plot(days, temp_min, label = 'Min', alpha = 0.3)
plt.plot(days, temp_max, label = 'Max', alpha = 0.3)
plt.legend(loc = 'best', shadow=True, fontsize='small')
plt.grid()
```
