### Tutorial 4 - Bar chart

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
company = ['Google', 'Meta', 'Ebay', 'Netflix']
revenue = [110, 92, 76, 89]
```

```python
xpos = np.arange(len(company))
xpos
```

```python
plt.xticks(xpos, company)
plt.xlabel('Company')
plt.ylabel('Revenue Billion')
plt.bar(xpos, revenue, label='Revenue')
plt.legend()
```

```python
profit = [70, 62, 48, 72]
plt.xticks(xpos, company)
plt.xlabel('Company')
plt.ylabel('Revenue/Profit Billion')
plt.bar(xpos, revenue, label='Revenue')
plt.bar(xpos, profit, label='Profit')
plt.legend()
```

```python
profit = [70, 62, 48, 72]
plt.xticks(xpos, company)
plt.xlabel('Company')
plt.ylabel('Revenue/Profit Billion')
plt.bar(xpos-0.2, revenue, label='Revenue')
plt.bar(xpos+0.2, profit, label='Profit')
plt.legend()
```

```python
profit = [70, 62, 48, 72]
plt.xticks(xpos, company)
plt.xlabel('Company')
plt.ylabel('Revenue/Profit Billion')
plt.bar(xpos-0.2, revenue, width=0.5, label='Revenue')
plt.bar(xpos+0.2, profit, width=0.5, label='Profit')
plt.legend()
```

```python
profit = [70, 62, 48, 72]
plt.xticks(xpos, company)
plt.xlabel('Company')
plt.ylabel('Revenue/Profit Billion')
plt.bar(xpos-0.2, revenue, width=0.5, label='Revenue')
plt.bar(xpos+0.2, profit, width=0.5, label='Profit')
plt.legend()
plt.grid()
```

```python
profit = [70, 62, 48, 72]
ypos = np.arange(len(company))
plt.yticks(ypos, company)
plt.xlabel('Revenue/Profit Billion')
plt.ylabel('Company')
plt.barh(ypos, revenue, label='Revenue')
plt.barh(ypos, profit, label='Profit')
plt.legend()
```

```python
ypos = np.arange(len(company))
profit = [70, 62, 48, 72]
plt.yticks(ypos, company)
plt.xlabel('Revenue/Profit Billion')
plt.ylabel('Company')
plt.barh(ypos-0.2, revenue, label='Revenue')
plt.barh(ypos+0.2, profit, label='Profit')
plt.legend()
```

```python
profit = [70, 62, 48, 72]
ypos = np.arange(len(company))
plt.yticks(ypos, company)
plt.xlabel('Revenue/Profit Billion')
plt.ylabel('Company')
plt.barh(ypos-0.2, revenue, height=0.5, label='Revenue')
plt.barh(ypos+0.2, profit, height=0.5, label='Profit')
plt.legend()
```
