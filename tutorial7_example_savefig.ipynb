### Tutorial 7 : Saving the chat using savefig

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
values = [750, 1050, 500]
cht_labels = ['India', 'Usa', 'China']
```

```python
plt.pie(values, labels=cht_labels, autopct='%0.0f%%')
plt.savefig('tutorial_7_myfile.png')
```

```python
plt.pie(values, labels=cht_labels, autopct='%0.0f%%')
plt.savefig('D:/matplotlib/tutorial_7_myfile2.pdf')
```

```python
plt.pie(values, labels=cht_labels, autopct='%0.0f%%')
plt.savefig('tutorial_7_myfile3.png', bbox_inches='tight')   # gives in box and fit size
```

```python
plt.pie(values, labels=cht_labels, autopct='%0.0f%%')
plt.savefig('tutorial_7_myfile4.png', bbox_inches='tight', pad_inches=3)
```

```python
plt.pie(values, labels=cht_labels, autopct='%0.0f%%')
plt.savefig('tutorial_7_myfile5.png', bbox_inches='tight', transparent = True)  # To create image with background transparent
```
