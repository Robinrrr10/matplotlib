## Tutorial 2 : Format string

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = [1, 2, 3, 4, 5, 6, 7]
y = [40, 39, 41, 43, 42, 40, 41]
```

```python
plt.plot(x, y, 'g+')
```

```python
plt.plot(x, y, 'g+--')
```

```python
plt.plot(x, y, 'b*--')
```

```python
plt.plot(x, y, 'D--')
```

```python
plt.plot(x, y, '*')
```

```python
plt.plot(x, y, color='yellow')
```

```python
plt.plot(x, y, color='#678823', marker='*', linestyle='dotted', markersize=13)
```

```python
plt.plot(x, y, color='blue', alpha=0.5)
```

```python
plt.plot(x, y, color='blue', alpha=0.1)
```
